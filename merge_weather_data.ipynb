```python
import pandas as pd
import datetime
```

```python
def filter_data(df):
    df = df[['city', 'year', 'month', 'day', 'temperature']]
    df = df[df['city'] == 'KRAKÃ“W-OBSERWATORIUM']
    return df
```

```python
START_YEAR = 2001
END_YEAR = 2021
START_MONTH = 1
END_MONTH = 8
```

```python
it_year = START_YEAR
merged_df = pd.DataFrame(columns=['city', 'year', 'month', 'day', 'temperature'])
while it_year <= END_YEAR:
    it_month = START_MONTH
    while it_month <= 12:
        if (it_year == 2021) and (it_month>END_MONTH):
            break
        if it_month < 10:
            file_prefix = 'k_d_t_0'
        else:
            file_prefix = 'k_d_t_'
        filename = f"{file_prefix}{it_month}_{it_year}.csv"
        print(f"Reading file: {filename}")
        df = pd.read_csv(filename, 
                         names=['a', 'city', 'year', 'month', 'day', 'temperature', 'b', 'c', 'd', 'e', 'f', 'g', 'h'], 
                         encoding="latin2")
        df = filter_data(df)
        merged_df = merged_df.append(df).reset_index(drop=True)
        
        it_month += 1
    
    it_year += 1
```
